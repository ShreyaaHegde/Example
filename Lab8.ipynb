```python
stplists=stopwords.words("english")
```

```python
stplists
```

```python
nltk_tokens=nltk.word_tokenize(str(data))
```

```python
nltk_tokens
```

```python
words_no_punc = []
```

```python
for word in nltk_tokens:
    if word.isalpha():
        words_no_punc.append(word)
```

```python
words_no_punc
```

```python
clean_words=[]
for word in words_no_punc:
    if word not in stplists:
        clean_words.append(word)
#clean_words
```

```python
newwords = " ".join(clean_words)
```

```python
newwords
```

```python
wc = WordCloud().generate(newwords)
```

```python
plt.imshow(wc)
```
